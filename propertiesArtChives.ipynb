```python
import pprint
import rdflib
from rdflib import URIRef, Literal, Namespace
from rdflib.namespace import XSD, RDFS, DCTERMS
from rdflib import Literal
```

```python
# create an empty Graph
g = rdflib.ConjunctiveGraph()

# parse a local RDF file by specifying the format into the graph
result = g.parse("../resources/artchives.nq", format='nquads')
```

```python
#tutte le properties di artchives
unique_prop = set()
for subj, pred, obj, context in g.quads():
    unique_prop.add(pred)
    
for pred in unique_prop:
    pprint.pprint(pred)
```

```python
from rdflib import Namespace

# assign prefixes to namespaces
wd = Namespace("http://www.wikidata.org/entity/") # remember that a prefix matches a URI until the last slash (or hashtag #)
wdt = Namespace("http://www.wikidata.org/prop/direct/")
art = Namespace("https://w3id.org/artchives/")
```

```python
#In ARTchives, all the topics (people, organisations, movements, genres) studied by (or somehow related to) historians 
#are linked to the historian by means of the predicate P921

unique_properties = set()

for s,p,o in g.triples((None, wdt.P921, None)):  #esempio di tutti gli oggetti di P921 (main subject (primary topic of a work)
    for s1,p1,o1 in g.triples((o, RDFS.label, None)):
        unique_properties.add(o1.strip())
        
for topic in unique_properties:
    print(topic)
```
